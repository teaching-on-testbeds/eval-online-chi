## Set up security groups

The rest of these materials assume security groups are already set up within the project! 

* If you are an instructor using these materials in the classroom, you should run this notebook to set up security groups on behalf of the entire class.
* If you are an individual working on these materials independently of a class assignment, run this notebook before you begin, to set up the security groups that don't already exist in your project.

```python
# runs in Chameleon Jupyter environment
import chi, os, time, datetime
from chi import context
from chi import network

context.version = "1.0" 
context.choose_project()
context.choose_site(default="KVM@TACC")
```

```python
# runs in Chameleon Jupyter environment
security_groups = [
  {'name': "allow-ssh", 'port': 22, 'description': "Enable SSH traffic on TCP port 22"},
  {'name': "allow-5000", 'port': 5000, 'description': "Enable TCP port 5000 (used by Flask)"},
  {'name': "allow-8000", 'port': 8000, 'description': "Enable TCP port 8000 (used by FastAPI)"},
  {'name': "allow-8888", 'port': 8888, 'description': "Enable TCP port 8888 (used by Jupyter)"},
  {'name': "allow-3000", 'port': 3000, 'description': "Enable TCP port 3000 (used by Grafana)"},
  {'name': "allow-9090", 'port': 9090, 'description': "Enable TCP port 9090 (used by Prometheus)"},
  {'name': "allow-8080", 'port': 8080, 'description': "Enable TCP port 8080 (used by cAdvisor, Label Studio)"}
]
```

```python
# runs in Chameleon Jupyter environment
for sg_def in security_groups:

    sg_list = network.list_security_groups(name_filter=sg_def["name"])
    if sg_list: # already exists
        pass
    else:       # create new security group
        sg = network.SecurityGroup({"name": sg_def["name"], "description": sg_def["description"]})
        sg.add_rule("ingress", "tcp", sg_def["port"])
        sg.submit()
```
